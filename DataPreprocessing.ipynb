```python


remaining = ['ano', 'est_civil', 'gru_ed1', 'gru_ed2', 'nivel_edu', 'ultcurfal', 'ocupacion', 'idpertet',
             'codptore', 'codmunre', 'area_res', 'seg_social', 'idadmisalud', 'asis_med', 'c_dir1', 'c_bas1',
             'c_mcm1', 'causa_667', 'cau_homol']
lst = list()
for ano in range(1998, 2008):
    print(ano)
    lst.append(utils.OrganizarDataFrame("defun/Defun" + str(ano) + ".sav"))

df_1998 = pd.concat(lst)
df_1998.columns = df_1998.columns.str.lower()
df_1998 = df_1998[(df_1998.cau_homol == "024") | (df_1998.cau_homol == "24") | (df_1998.cau_homol == 24)]

lst = list()
for ano in range(2008, 2019):
    print(ano)
    lst.append(utils.OrganizarDataFrame("defun/Defun_" + str(ano) + ".sav"))

df_2008 = pd.concat(lst)
df_2008.columns = df_2008.columns.str.lower()
df_2008 = df_2008[(df_2008.cau_homol == "024") | (df_2008.cau_homol == "24")]

```

```python
###
borrar = list(set(df_1998.columns) - set(remaining).intersection(set(df_1998.columns)))
borrar = set(borrar)
borrar2 = list(set(df_2008.columns) - set(remaining).intersection(set(df_2008.columns)))
borrar2 = set(borrar2)

df_1998 = df_1998.drop(columns=borrar)
df_2008 = df_2008.drop(columns=borrar2)
df = pd.concat([df_1998, df_2008])
df
```

```python
for col_name in df.columns:
    utils.remove_blanks(df,col_name)

df
```
